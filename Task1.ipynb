### **Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population. (Adult Income Dataset)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('adult.csv')
```

```python
data1
```

```python
g_counts = data1['gender'].value_counts()
plt.bar(g_counts.index, g_counts, color=['skyblue', 'yellow'], edgecolor="black")
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender in Adult Income Dataset')
plt.show()
```

```python
sns.histplot(data1['age'], bins=20, kde=True, color="blue")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age in Adult Income Dataset')
plt.show()
```

**Bar Plot for Age vs Income**

```python
sns.barplot(x='age', y='income', data=data1, color="pink")
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.show()
```

**Bar Plot for Education vs Income**

```python
sns.countplot(x='education', hue='income', data=data1)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Education vs Income')
plt.legend(title='Income', loc='upper right', labels=['<=50K', '>50K'])
plt.show()
```

**Histogram for distribution of education-num and hours-per-week**

```python
features = ['educational-num', 'hours-per-week']
plt.figure(figsize=(14, 5))
for i, feature in enumerate(features, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data1[feature].dropna(), bins=20, kde=True, color='green')
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```
